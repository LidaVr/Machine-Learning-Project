The purpose of this script is to rename all of the images so that they share the same naming convention. For each category (already 7 folders) we rename the images as such: {class_name}_{4 digit image_number}.{image format}

```python
#Connect to GoogleDrive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Installing the necessary libraries
import os
import shutil
```

```python
#Set-up the directory to the assignment folder
os.chdir('/content/drive/My Drive/Colab Notebooks')
```

```python
#Function to rename the images according to the required convention.
def rename_and_move_images(source_dir, destination_dir):
    classes = os.listdir(source_dir)

    for class_name in classes:
        class_dir = os.path.join(source_dir, class_name)   #Create output folder if it does not already exist
        if not os.path.isdir(class_dir):
            continue

        #Create all class subfolders in the destination directory
        class_destination_dir = os.path.join(destination_dir, class_name)
        os.makedirs(class_destination_dir, exist_ok=True)

        files = os.listdir(class_dir)

        for idx, file_name in enumerate(files):
            file_format = os.path.splitext(file_name)[-1]
            new_file_name = f"{class_name}_{str(idx).zfill(4)}{file_format}"
            old_file_path = os.path.join(class_dir, file_name)
            new_file_path = os.path.join(class_destination_dir, new_file_name)
            shutil.copy(old_file_path, new_file_path) #Copy images to new destination, do not replace in old one
```

```python
#Set-up of source and destination directories
source_dir = "Extracted_Images"
destination_dir = "Renamed_Images"
```

```python
rename_and_move_images(source_dir, destination_dir)
```
